```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loaded in filepath 
filepath = 'N_12_extent_v3.0 (1).csv'
df = pd.read_csv(filepath)

# Changing the range from 2000 - 2022 because there was a weird error in 1987 that throws off the data. 
df = df.drop(index=range(0, 22))
df = df.reset_index(drop=True)

#display(df)

# Making graph 1. This is data from the N. Arctic sea ice extent from 2000-2022
fix, ax = plt.subplots(figsize = (15,5))

year = df['year']

# Note for later: There is weird spacing in front of the column names
extent = df[' extent']

ax.plot(year, extent, c = 'blue', linestyle = '-', marker = '.', markersize = '12', label = 'Arctic Sea Ice Extent')
ax.set_title('N. Arctic Sea Ice Extent in December (2000 - 2022)', fontsize = '20')
ax.set_ylabel('Sea Ice Extent (millions of square kilometers)', fontsize = '15')
ax.set_xlabel('Year', fontsize = '15')
ax.grid(linestyle = '-')
ax.legend(loc = 'upper right')

# Add trend line
z = np.polyfit(year, extent, 1)
p = np.poly1d(z)
plt.plot(year, p(year), color = 'fuchsia', linewidth = 3, linestyle = '--')

plt.show()

```

```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt

# Loading in the filepath
file_path = '/home/jovyan/Milestone 3/N_Sea_Ice_Index_Regional_Monthly_Data_G02135_v3.0 (3).xlsx'
df = pd.read_excel(file_path)

# Only using years 2000-2022
df = df.drop(index=range(0, 24))
df = df.reset_index(drop=True)

# Dropping the year 2023 for consistency
df = df.drop(index = 23)

# Printing column names
print(df.columns)

# Removing the ranks which are all 'Unnamed: #' that are after 0. Got this info from the excel sheet that I can see, just not in Jupyter. Got help from stack overflow.
df = df.iloc[:, [0] + list(range(1, df.shape[1], 2))]

# Displaying dataframe 
display(df)

# Extracting months and area index values
months = df.columns[1:]  
area_index = df.iloc[:, 1:] 

# Transposing the DataFrame
df_transposed = df.transpose()
display(df_transposed)

# Simplifying the years so it is not a ton of things plotted
years_wanted = [2000.0, 2005.0, 2010.0, 2015.0, 2020.0, 2022.0]

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))
column_names = df.columns

# I used a for loop to loop through each column. I had to include a year label so that it would plot in the legend correctly. 
for column in df_transposed.columns[0:]:
    year_label = df_transposed.iloc[0, column]
    if year_label in years_wanted:
        ax.plot(df_transposed.index[1:], df_transposed[column][1:], label = year_label) 

# Adding labels and title
ax.set_xlabel('Month', size = 15)
ax.set_ylabel('Sea Ice Extent (Million Km^2)', size = 15)
ax.set_title('N. Arctic Sea Ice Index (2000-2022)', size = 18)

# Adding the legend
ax.legend(title='Year', loc = 'upper right', bbox_to_anchor = (1.10,1.05))
ax.grid(linestyle = '-')
plt.show()










```
